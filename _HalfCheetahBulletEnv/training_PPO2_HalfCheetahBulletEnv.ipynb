````
env = gym.make("HalfCheetahBulletEnv-v0")

model = PPO2('MlpPolicy', env, gamma=0.99, n_steps=2048, ent_coef=0.0, 
               learning_rate=3e-4, lam=0.95, nminibatches=32, noptepochs=10,
               cliprange=0.2, verbose=1)

model.learn(total_timesteps=int(2e6))
````

````
# Create save dir
save_dir = "saves/"
os.makedirs(save_dir, exist_ok=True)

model.save(save_dir + "/PPO2_HalfCheetahBulletEnv")
````
